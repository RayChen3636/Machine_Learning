```python
from keras.datasets.mnist import load_data
# ((x_train, y_train), (x_test, y_test))
(x_train, y_train), (x_test, y_test) = load_data()
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import random
i = random.randint(0, 59999)
plt.title("This is {}".format(y_train[i]))
plt.imshow(x_train[i], cmap="gray")
```

```python
import pandas as pd
pd.DataFrame(x_train[i])
```

```python
# 剩一個位置打-1, python 會幫你算剩下的
# pd.DataFrame(x_train.reshape( -1, 28 * 28))
x_train_reshape = x_train.reshape( -1, 28 * 28)
x_test_reshape = x_test.reshape( -1, 28 * 28)
```

```python
# n_componets = 30 篩選的最大值
from sklearn.decomposition import PCA
pca = PCA(n_components=30)
x_train_pca = pca.fit_transform(x_train_reshape)
x_test_pca = pca.transform(x_test_reshape)
x_test_pca
```

```python
# 新的座標軸對於原始座標軸的關係
pd.DataFrame(pca.components_)
```

```python
# 畫圖有兩種選擇: 0-255整數 or 0-1小數
newaxis = pca.components_.reshape(-1, 28, 28)
# import numpy as np
# 將值 normalize 到 0-1小數, 但 python matplotlib, 已經做過了
# newaxis = (newaxis -np.min(newaxis))/np.ptp(newaxis), ptp為 range of value(maximum - minimum) along an axis
width = 5
height = newaxis.shape[0] // width + 1
plt.figure(figsize=(10, 14))
for i, a in enumerate(newaxis):
    plt.subplot(height, width, i+1)
    plt.axis("off")
    plt.title("Axis:{}".format(i+1))
    plt.imshow(a, cmap="gray")
newaxis.shape
```

```python
ratio = pca.explained_variance_ratio_
print("比例:", ["{}%".format(round(r*100, 2)) for r in ratio])
print("比例加總:", "{}%".format(sum(ratio)*100))
# print("數值:", pca.explained_variance_)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV

clf = RandomForestClassifier()
params = {
    "n_estimators":range(0, 150, 10),
    "max_depth":[5, 15, 20]
}
grid = GridSearchCV(clf, params, cv=10, n_jobs=6)
grid.fit(x_train_pca, y_train)
print("最佳參數:", grid.best_params_)
print("最佳分數:", grid.best_score_)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import RandomForestClassifier
import numpy as np
clf = RandomForestClassifier(n_estimators=100, max_depth=20)
scores = cross_val_score(clf, x_train_pca, y_train, cv=10, n_jobs=6)
print("十次分數:", scores)
print("平均分數:", np.average(scores))
```

```python
clf = RandomForestClassifier(n_estimators=100, max_depth=20)
clf.fit(x_train_pca, y_train)
```

```python
from sklearn.metrics import accuracy_score
pre = clf.predict(x_test_pca)
accuracy_score(pre, y_test)
```

```python
from sklearn.metrics import confusion_matrix
mat = confusion_matrix(y_test, pre)
result = []
for r in mat:
    temp = ["{}%".format(round(n/sum(r) * 100), 2) for n in r]
    result.append(temp)
pd.DataFrame(result,
             index=["{}(真實)".format(r) for r in range(10)],
             columns=["{}(預測)".format(r) for r in range(10)])
```
