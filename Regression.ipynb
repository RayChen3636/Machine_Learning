```python
# 從 scikit learn, 載入波士頓房價資料
from sklearn.datasets import load_boston
import pandas as pd
boston = load_boston()
df = pd.DataFrame(boston["data"], columns=boston["feature_names"])
df["target"] = boston["target"]
df
```

```python
# (有看過的資料)做出公式 -> (沒看過的資料)驗證一下公式的好壞
from sklearn.model_selection import train_test_split
# [90%題目, 10%題目, 90%答案, 10%答案]
x_train, x_test, y_train, y_test= train_test_split(boston["data"], 
                                                   boston["target"], 
                                                   test_size=0.1)
```

```python
# 做公式(fit), 即訓練資料, 有答案的訓練即監督式學習
from sklearn.tree import DecisionTreeRegressor
reg = DecisionTreeRegressor(max_depth=5)
reg.fit(x_train, y_train)
```

```python
import graphviz
from sklearn.tree import export_graphviz
g = export_graphviz(reg, out_file=None, 
                    feature_names=boston["feature_names"], 
                    filled=True)
graphviz.Source(g)
```

```python
# 驗證一下公式的好壞
# 預測(predict)
print("預測結果:", list(reg.predict(x_test)))
print("正確結果:", list(y_test))
from sklearn.metrics import r2_score
print("r2 score:", r2_score(y_test, reg.predict(x_test)))
```
